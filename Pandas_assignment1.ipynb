```python
# 3.Write a Python program to add, subtract, multiple and divide two Pandas Series.
import pandas as pd
series1=pd.Series([1,2,3,4,5])
series2=pd.Series([6,7,8,9,10])
print(series1+series2)
print(series1-series2)
print(series1*series2)
print(series1/series2)
```

```python
# 4 .Write a Python program to compare the elements of the two Pandas Series.
import pandas as pd
series1=pd.Series([1,2,3,9,10])
series2=pd.Series([6,7,8,9,5])
print('Equals to', series1==series2)
print('greater than', series1>series2)   
print('less than', series1<series2)

```
